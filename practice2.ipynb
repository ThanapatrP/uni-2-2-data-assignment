# 6. นำคอลัมน์ที่ชื่อ continent ของ drinks ไปต่อกับตาราง medals ตามชื่อประเทศของ medals
1. การที่จะ join ตารางได้นั้นต้องมีคอลัมน์ที่ตรงกันในทั้งสองตาราง ในที่นี้ให้ใช้ชื่อประเทศ
2. อย่าลืมว่าการ join ข้อมูลนั้นเราต้องการคงข้อมูลของประเทศที่มีในตาราง medals ไว้ ตารางที่ join
แล้วจะมีจำนวนแถวเท่ากับตาราง medals

```python
medals = medals.join(drinks.set_index("country").loc[:, "continent"], on="Country")
medals
```

# 7. ในตาราง medals จงแสดงชื่อประเทศที่ยังไม่มีข้อมูลในคอลัมน์ continent


```python
medals[medals.continent.isna()]
```

# 8. นิสิตจะเห็นแนวทางการแก้ไขข้อมูลหรือรวมประเทศบางประเทศเข้าด้วยกัน แต่ในการบ้านนี้ยังไม่ต้องทำทั้งหมด แค่เติมข้อมูลทวีปที่ยังขาดหายไปในบางประเทศดังนี้ Australasia: OC, Canada/United States/Mexico: NA, Cuba/Jamaica: SA, Czech/Yugoslavia/Germany(ทั้งหมด)/Great Britain/Russia(ทั้งหมด)/Soviet: EU ประเทศที่เหลือกำหนดให้เป็น 'OT' ไป
1. ตารางนี้จะเป็นตารางแบบ wide ที่สมบูรณ์ ควรเก็บไว้เพื่อใช้ในการคำนวณ

```python
medals[medals.Country.str.contains("Germany")]
```

```python
medals[medals.Country.str.contains("Russia")]
```

```python
medals[medals.Country.str.contains("Soviet")]
```

```python
medals.loc[medals.Country == "Australasia", "continent"] = "OC"
medals.loc[medals.Country.isin(["Canada", "United States", "Mexico"]), "continent"] = "NA"
medals.loc[medals.Country.isin(["Cuba", "Jamaica"]), "continent"] = "SA"
medals.loc[medals.Country.isin(["Czech", "Yugoslavia", "Germany", "United Team of Germany", "East Germany"
                                , "West Germany", "Great Britain", "Russia", "Russian Empire", "Soviet Union"]), "continent"] = "EU"
medals.loc[medals.continent.isna(), "continent"] = "OT"
```

```python
medals[medals.continent.isna()]
```

# 9. ตั้งให้ ชื่อทวีปเป็น index ชั้นนอกและชื่อประเทศเป็น index ชั้นใน

```python
medals.set_index(["continent", "Country"], inplace=True)
medals
```

# 10. เรียงลำดับชื่อทวีปจากมากไปน้อยและในทวีปเรียงลำดับชื่อประเทศจากน้อยไปมาก

```python
medals.sort_index(ascending = [False, True], inplace=True)
medals
```

# 11. จัดคอลัมน์ให้เป็นสองระดับ โดยระดับแรกคือฤดู และระดับที่สองคือจำนวนเกมส์และเหรียญที่ได้ (ดูตัวอย่างจากเอกสารการสอน) ดังนั้นตอนนี้ข้อมูลจะมีสองระดับทั้ง index (ทวีปและประเทศ) และ column(ฤดูและจำนวนเกมและเหรียญ)

```python
new_cols = [ ("Code", ""),
        ("Summer", "SummerGame"),
        ("Summer", "SummerGold"),
        ("Summer", "SummerSilver"),
        ("Summer", "SummerBronze"),
        ("Winter", "WinterGame"),
        ("Winter", "WinterGold"),
        ("Winter", "WinterSilver"),
        ("Winter", "WinterBronze"),
        ]
```

```python
old_cols = medals.columns
old_cols
```

```python
medals.columns = pd.MultiIndex.from_tuples(new_cols)
medals
```

# 12. ย้ายระดับของฤดูมาเป็น index ชั้นในสุด

```python
medals.swaplevel(0, 1,axis="columns")
```

# 13. ปรับตารางให้เป็นแบบ long โดยหนึ่งแถวจะมีคอลัมน์คือทวีป ประเทศ ฤดูแข่งขัน จำนวนเกม ประเภทเหรียญ จำนวนเหรียญ

```python
medals.columns = old_cols
medals = medals.reset_index()
medals
```

```python
medals = medals.melt(id_vars=['continent', 'Country', 'SummerGame', 'WinterGame'],
                     value_vars=['SummerGold', 'SummerSilver', 'SummerBronze', 
                    'WinterGold', 'WinterSilver', 'WinterBronze'],
                    var_name='Medal', 
                    value_name='Count')
```

```python
medals["Season"] = medals["Medal"].str.extract("(Summer|Winter)")
medals["Medal"] = medals["Medal"].str.replace("(Summer|Winter)", "", regex=True)
medals["Game"] = np.where(medals["Season"] == "Summer",
                          medals["SummerGame"],
                          medals["WinterGame"]
                          ) # (condition, if true, if false)
```

```python
medals = medals[['continent', 'Country', 'Season', 'Game', 'Medal', 'Count']]
```

```python
medals.head(10)
```

# 14. แสดงจำนวนเหรียญทองโอลิมปิกฤดูร้อนรวมของแต่ละทวีปจากมากไปน้อย

```python
summer_gold = medals[(medals.Season == 'Summer') & (medals.Medal == 'Gold')]

x = summer_gold.groupby('continent')

x.groups
```

```python
x.Count.sum().sort_values(ascending=False)
```

# 15. นับว่าแต่ละทวีปมีกี่ประเทศ (ลองใช้คำสั่ง size() ต่อจากการ groupby() )

```python
x.Country.size()
```

# 16. แสดงตารางรวมเหรียญฤดูร้อนแต่ละประเภทของทุกประเทศในทวีปเดียวกัน แสดงผลแยกตามทวีป


```python
summer = medals[medals['Season'] == 'Summer'] # filter season == summer

continent_medals = summer.groupby(['continent', 'Medal']) #groupby continent and medal

continent_medals = continent_medals.Count.sum()
continent_medals
```

```python
x = continent_medals.unstack()

x = x[['Gold', 'Silver', 'Bronze']]
x.columns = ['Summer' + col for col in x.columns]

x
```

# 17. แสดงตารางสรุปรวมเหรียญแต่ละประเภทของทุกประเทศแยกตามฤดูแข่งขัน

```python
x = medals.groupby(['Medal', 'Season'])
x.Count.sum().unstack()
```

# 18. หาค่าเฉลี่ยของจำนวนเหรียญทองฤดูร้อนต่อประเทศที่เข้าร่วมแข่งขัน แยกตามทวีป 

```python
summer_gold = medals[(medals.Season == 'Summer') & (medals.Medal == 'Gold')]

avg_gold = summer_gold.groupby('continent')['Count'].mean()

avg_gold
```

# 19. หาจำนวนประเทศที่เคยเข้าร่วมการแข่งขันโอลิมปิค แยกตามทวีปและฤดูแข่งขัน แสดงในตารางเดียวกัน


```python
x = medals[medals.Game > 0]
x = x.groupby(['continent', 'Season'])
x.Country.nunique().unstack()
```
