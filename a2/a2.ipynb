```python
import pandas as pd
import numpy as np
```

# 1. โหลดไฟล์ที่ชื่อ olympics_cleaned.csv ขึ้นมาใน DataFrame ที่ชื่อ medals

```python
medals = pd.read_csv("olympics_cleaned.csv")
medals
```

# 2. โหลดไฟล์ชื่อ drinks.csv ขึ้นมาใน DataFrame ชื่อ drinks

```python
drinks = pd.read_csv("drinks.csv")
drinks
```

# 3. ใน drinks มีคอลัมน์ไหนที่ไม่มีข้อมูลบ้าง และไม่มีข้อมูลกี่ตำแหน่งในแต่ละคอลัมน์

```python
drinks.isna().sum()
```

# 4. ลองเช็คว่าชื่อประเทศของตาราง medals ใดที่ไม่มีอยู่ในตาราง drinks 
    1. ลองใช้คำสั่ง isin() ในการเช็คว่ามีประเทศใดใน medals ไม่อยู่ใน drinks บ้าง

```python
medals[medals.Country.isin(drinks.country) == False]
```

# 5. แสดงชื่อประเทศของตาราง drinks ที่ไม่มีอยู่ในตาราง medals 
    1. นิสิตจะเห็นความแตกต่างของชื่อประเทศในสองตารางนี้ว่าเรียกชื่อประเทศไม่เหมือนกัน ตอนทำงาน

```python
drinks[drinks.country.isin(medals.Country) == False]
```

# 6. นำคอลัมน์ที่ชื่อ continent ของ drinks ไปต่อกับตาราง medals ตามชื่อประเทศของ medals 
    1. การที่จะ join ตารางได้นั้นต้องมีคอลัมน์ที่ตรงกันในทั้งสองตาราง ในที่นี้ให้ใช้ชื่อประเทศ 
    2. อย่าลืมว่าการ join ข้อมูลนั้นเราต้องการคงข้อมูลของประเทศที่มีในตาราง medals ไว้ ตารางที่ join 
แล้วจะมีจำนวนแถวเท่ากับตาราง medals

```python
medals = medals.join(drinks.set_index("country").loc[:,"continent"], on='Country')
medals
```

# 7. ในตาราง medals จงแสดงชื่อประเทศที่ยังไม่มีข้อมูลในคอลัมน์ continent

```python
medals[medals.continent.isna() == True]
```

# 8. นิสิตจะเห็นแนวทางการแก้ไขข้อมูลหรือรวมประเทศบางประเทศเข้าด้วยกัน แต่ในการบ้านนี้ยังไม่ต้องทำทั้ง หมด แค่เติมข้อมูลทวีปที่ยังขาดหายไปในบางประเทศดังนี้ Australasia: OC, Canada/United States/ Mexico: NA, Cuba/Jamaica: SA, Czech/Yugoslavia/Germany(ทั้งหมด)/Great Britain/Russia(ทั้ง หมด)/Soviet: EU ประเทศที่เหลือกำหนดให้เป็น 'OT' ไป 
1. ตารางนี้จะเป็นตารางแบบ wide ที่สมบูรณ์ ควรเก็บไว้เพื่อใช้ในการคำนวณ

```python
medals.loc[medals.Country == "Australasia", "continent"] = "OC"
medals.loc[medals.Country.isin(["Canada", "United States", "Mexico"]), "continent"] = "NA"
medals.loc[medals.Country.isin(["Cuba", "Jamaica"]), "continent"] = "SA"
medals.loc[medals.Country.str.contains("Czech\b|Yugoslavia|Germany|Great Britain|Russia|Soviet", regex=True), "continent"] = "EU"
medals.loc[medals.continent.isna(), "continent"] = "OT"
```

# 9. ตั้งให้ ชื่อทวีปเป็น index ชั้นนอกและชื่อประเทศเป็น index ชั้นใน

```python
medals.set_index(["continent", "Country"], inplace=True)
medals
```

# 10. เรียงลำดับชื่อทวีปจากมากไปน้อยและในทวีปเรียงลำดับชื่อประเทศจากน้อยไปมาก

```python
medals.sort_index(ascending=[False, True], inplace=True)
medals
```

# 11. จัดคอลัมน์ให้เป็นสองระดับ โดยระดับแรกคือฤดู และระดับที่สองคือจำนวนเกมส์และเหรียญที่ได้ (ดู ตัวอย่างจากเอกสารการสอน) ดังนั้นตอนนี้ข้อมูลจะมีสองระดับทั้ง index (ทวีปและประเทศ) และ column (ฤดูและจำนวนเกมและเหรียญ)

```python
columns = pd.MultiIndex.from_tuples([
    ("Code", ""),
    ("Summer", "SummerGame"),
    ("Summer", "SummerGold"),
    ("Summer", "SummerSilver"),
    ("Summer", "SummerBronze"),
    ("Winter", "WinterGame"),
    ("Winter", "WinterGold"),
    ("Winter", "WinterSilver"),
    ("Winter", "WinterBronze"),
    ])

old_col = medals.columns

medals.columns = columns
medals
```

# 12. ย้ายระดับของฤดูมาเป็น index ชั้นในสุด

```python
medals.swaplevel(i=0, axis=1)
```

# 13. ปรับตารางให้เป็นแบบ long โดยหนึ่งแถวจะมีคอลัมน์คือทวีป ประเทศ ฤดูแข่งขัน จำนวนเกม ประเภทเหรียญ จำนวนเหรียญ

```python
medals.columns = old_col
medals = medals.reset_index()
```

```python
melted = pd.melt(medals,
                 id_vars=["continent", "Country", "SummerGame", "WinterGame"],
                 value_vars=["SummerGame", "SummerGold", "SummerSilver", "SummerBronze", "WinterGame", "WinterGold", "WinterSilver", "WinterBronze"],
                 var_name = "Medal",
                 value_name="Count")
melted["Season"] = melted.Medal.str.extract("(Summer|Winter)")
melted["Game"] = np.where(melted["Medal"].str.contains("Summer"), melted["SummerGame"], melted["WinterGame"])
bak_melt = melted.copy() # back up for before rename column of the dataframe
melted["Medal"] = melted["Medal"].str.replace("(Summer|Winter)", "", regex=True)
melted = melted.loc[:, ["continent", "Country", "Season", "Game", "Medal", "Count"]]
melted
```

# 14. แสดงจำนวนเหรียญทองโอลิมปิกฤดูร้อนรวมของแต่ละทวีปจากมากไปน้อย

```python
melted = bak_melt
melted[melted["Medal"] == "SummerGold"][["continent", "Count"]].groupby("continent").sum().sort_values(by="Count", ascending=False)
```

# 15. นับว่าแต่ละทวีปมีกี่ประเทศ (ลองใช้คำสั่ง size() ต่อจากการ groupby() )

```python
medals.groupby("continent").size()
```

# 16. แสดงตารางรวมเหรียญฤดูร้อนแต่ละประเภทของทุกประเทศในทวีปเดียวกัน แสดงผลแยกตามทวีป

```python
sum_medal_continent = melted[melted["Season"] == "Summer"][["continent", "Medal", "Count"]].groupby(["continent", "Medal"]).sum()
sum_medal_continent = sum_medal_continent.unstack("Medal").droplevel(axis=1, level=0).reset_index()
sum_medal_continent.columns.name = None

sum_medal_continent = sum_medal_continent.set_index("continent")
sum_medal_continent[["SummerGold", "SummerSilver", "SummerBronze"]]
```

# 17. แสดงตารางสรุปรวมเหรียญแต่ละประเภทของทุกประเทศแยกตามฤดูแข่งขัน

```python
melted["Medal"] = melted["Medal"].replace("(Summer|Winter)", "", regex=True)
sum_medal_season = melted[["Season", "Medal", "Count"]].groupby(["Season", "Medal"]).sum().unstack(level=0).droplevel(axis=1, level=0).reset_index().set_index("Medal")
sum_medal_season.columns.name = None
sum_medal_season
```

# 18. หาค่าเฉลี่ยของจำนวนเหรียญทองฤดูร้อนต่อประเทศที่เข้าร่วมแข่งขัน แยกตามทวีป

```python
melted
melted.reset_index()
melted_mean = melted[(melted["Season"] == "Summer") & (melted["Medal"] == "Gold")][["continent", "Count"]].groupby("continent").mean()
melted_mean
```

# 19. หาจำนวนประเทศที่เคยเข้าร่วมการแข่งขันโอลิมปิค แยกตามทวีปและฤดูแข่งขัน แสดงในตารางเดียวกัน

```python
medals["Summer"] = np.where(medals["SummerGame"] > 0, 1, 0)
medals["Winter"] = np.where(medals["WinterGame"] > 0, 1, 0)
medals[["continent", "Summer", "Winter"]].groupby("continent").sum()
```

# 20. จำนวนเหรียญรวม(ฤดูร้อนกับฤดูหนาว) แยกตามทวีปและฤดูแข่งขัน

```python
medals["Gold"] = medals["SummerGold"] + medals["WinterGold"]
medals["Silver"] = medals["SummerSilver"] + medals["WinterSilver"]
medals["Bronze"] = medals["SummerBronze"] + medals["WinterBronze"]

medals[["continent", "Bronze", "Gold", "Silver"]].groupby("continent").sum()
```
